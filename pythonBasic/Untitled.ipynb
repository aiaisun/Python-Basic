```python
url = "https://tabelog.com/tw/tokyo/rstLst/{}/?SrtT=rt"
```

```python
import pandas as pd
df = pd.DataFrame(columns=["評分","日文","英文"])
```

```python
for page in range(1,11):
    pageUrl = url.format(page)
    res = urlopen(pageUrl)
    html = BeautifulSoup(res)
    rs = html.find_all("li", class_ = "list-rst")
    
#     if page<10:
    for r in rs:
        JN = r.find("small", class_= "list-rst__name-ja")
        en = r.find("a", class_= "list-rst__name-main")
        rating = r.find("b", class_="c-rating__val")
        price = r.find_all("span", class_="c-rating__val")
        
        rData= {
            "Name"    : en.text,
            "JName"   : JN.text,
            "Website": en["href"],
            "Rate"   : rating.text,
            "Launch Price": price[0].text.replace("-","Close"),
            "Dinner Price": price[1].text.replace("-","Close")
        }
        #插入
        row = {"評分": rating.text, "日文": JN.text, "英文": en.text}
        #記得設定回去
        #沒有列標籤東西，append進去ignore_indes=true
        df = df.append(row, ignore_index = True,)
        print(en.text)
        print(JN.text)
        print("Website:", en["href"])
        print("Rate:", rating.text)
        print("Launch:", price[0].text.replace("-","Close"))
        print("Dinner:", price[1].text.replace("-","Close"))
        print("-"*100)
        
    print(pageUrl, "OK")

df.to_csv("tablelog.csv",encoding = "utf-8")
```
