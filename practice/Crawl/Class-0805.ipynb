```python
import requests, time, random
from bs4 import BeautifulSoup as bs

def getParsedData(soup,url):
#     標題 作者 時間
    news2 = soup.select("header.td-post-title")[0]
    title = news2.select("h1")[0].text
    author = news2.select("div.td-post-author-name")[0].text.replace("綜合報導","").replace(" ","").replace("-","").replace("/","").replace("\n","")
    time = news2.select("time")[0].text

#     內容    
    pTag = soup.select("span[itemprop='articleBody'] p")
    content=""
    for ele in pTag[:-1]:
        content+= ele.text
        content+="\n"
        
#     關鍵字        
    aTags = soup.select("div.td-post-source-tags li a")
    keyWords = [a.text for a in aTags]
    keyWordsList = ",".join(keyWords)
    
#     更多連結
    moreTag = soup.select("div.relativeArticle_Reynold li a")
    moreList = [ {"Title": ele.text.strip(), "Link": ele["href"]} for ele in moreTag]

    data = {
        "url"      : url,
        "Title"    : title,
        "Author"   : author,
        "Time"     : time,
        "Content"  : content,
        "Key Words": keyWordsList,
        "More"     : moreList
        
    }
    return data



base_url = "https://www.nownews.com/"
headers={
    "user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36"
}
res = requests. get(base_url,headers = headers)
soup = bs(res.text,"lxml")
news = soup.select("div.td-module-container")

base_link=[ele.select("a")[1]["href"] for ele in news]
# print(base_link)

dataList = []
for link in base_link:
    time.sleep(random.random()*2)
    url2 = link
    res2 = requests. get(url2, headers = headers)
    soup2 = bs(res2.text,"lxml")
    result = getParsedData(soup2,link)
    dataList.append(result)
    print("{} is done.".format(link))
    #     dataList += result XX

dataList


```

```python
with open("0805.csv","w", encoding="Big5") as outFile:
    outFile.write("時間,作者,標題,連結\n")
    
    for ele in dataList:
        content = (ele["Time"] + "," + ele["Author"] + "," + ele["Title"] + "," + ele["url"] + "\n")
        outFile.write(content)
```
