```python
import requests

url = "https://www.nownews.com/"

res = requests.get(url)

res.text
```

```python
from bs4 import BeautifulSoup as bs

soup = bs(res.text, "lxml")
print(soup)

```

```python
result = soup.select("div.tdb_module_loop h3")
for ele in result:
    print(ele)
    aTag = ele.select("a")[0]
    print(aTag)
    print("Link:", aTag["href"])
    print("Title:", aTag["title"])
    print("Content: ", aTag.text)
#     print(ele.find("a"))
#     print(type(ele))
    print("-"*80)
```
