```python
import requests

url = "https://www.nownews.com/"

res = requests.get(url)
res.text
```

```python
from bs4 import BeautifulSoup as bs

origin = bs(res.text,"lxml")
print(origin)
```

```python
point1 = origin.select("div.td-block-span6 h3")
# print(point1)

for ele in point1:
    aTag = ele.select("a")[0]
#     print(aTag)
    print("Title: ", aTag["title"])
    print("Link: ", aTag["href"])
    print("Content: ", aTag.text)
    print("-"*100)
    
```

```python
DB = []
for ele in point1:
    bTag = ele.select("a")[0]
    data={
        "Title:": bTag["title"],
        "Link: ": bTag["href"],
        "Content:": bTag.text
    }
    DB.append(data)

for ele in DB:
    print(ele)
    
```
