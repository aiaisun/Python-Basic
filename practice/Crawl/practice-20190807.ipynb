```python
import requests, random, time
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def getParsedData(link):
    headers={
    "user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36"
    }
    res = requests.get(link, headers = headers)
    soup = bs(res.text,"lxml")

    # 標題
    title = soup.select("div.articlebody h1")[0].text.replace("\n","")

    # 時間
    time = soup.select("div.text span")[0].text

    # 文章內容
    pTags = soup.select("div.text p")
    content = ""
    for ele in pTags[:-1]:
        content += ele.text
        content +="\n"
    
    data = {
        "Title"   : title,
        "Time"    : time,
        "Content" : content
    }
    

    return data
```

```python
def saveLtnFile(dataList,catagory):
    df = pd.DataFrame(dataList)
    df.to_excel("ltn_{}.xlsx".format(catagory), index = False)
```

```python
headers={
    "user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36"
}

target = ["world","society"]
catagory_url= "https://news.ltn.com.tw/list/breakingnews/{}"

for catagory in target:
    
    allNewsLink = []
    
    
    # 取得所有類別分頁  catagoryLink
    catagoryLink = []
    for catagoryPage in range(1,2):
        catagoryLink.append(catagory_url.format(catagory)+"/{}".format(catagoryPage))

        # 發request 取得頁面資料
#         catagoryLink= [1,2,3]
        pageNewsLink = []
        for page in catagoryLink:
            
            res2 = requests.get(page,headers = headers)
            soup2 = bs(res2.text,"lxml")

            # 取得文章列表 allNewsLink
            pageNewsLink = [ele["href"] for ele in soup2.select("div.whitecon li a.tit")]
        
#             pagenewslink(第一頁的20)
        revisedPageNewsLink = []
        for i in pageNewsLink:
            
            revisedPageNewsLink.append(i.replace("//","https://"))
          
        allNewsLink += revisedPageNewsLink
        print(page,"is done.")
  
    
# 爬所有網頁連結的資料,並存擋    
    allDataList = []
    for link in allNewsLink:
        
        time.sleep(random.random()*2)
        newsData = getParsedData(link)
        allDataList.append(newsData)
        print("{} is done.".format(link))
    try:
        saveLtnFile(allDataList,catagory)
        print("{} file saved".format(catagory))
    except:
        print("{} file fail saved".format(catagory))
```
