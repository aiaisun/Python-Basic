```python
import requests

# 使用API得到展覽頁面的soup_article

def getExhibitionSoup(API):
    headers = {
    "Cookie": "_ga=GA1.2.1313403307.1570685267; ASP.NET_SessionId=dto1xrxzjglq35bcyhcbbbtr; _gid=GA1.2.1014102315.1571625373",
    "Host": "www.songshanculturalpark.org",
    "Referer": "https://www.songshanculturalpark.org/",
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36",
    "X-CSRF-TOKEN": "pGKHRfaZOON8cyaKlR452FgaubvNc7C6nA6bx5vg",
    "X-Requested-With": "XMLHttpRequest"
    }
    exhibitionSoup = requests.get(API, headers = headers)
    return exhibitionSoup

# 得到展覽資訊

def getExhibitionInfo(soup_article):
    data = soup_article.json()
    exhibitionInfo = data["items"]
    
    exhibitionData = [{
        "Title"         : e["Title"],
        "PublishDate"   : e["PublishDate"],
        "Place"         : "松菸",
        "shortInfo"     : e["SubTitle"],
        "ID"            : e["ID"],
        "link"          : "https://www.songshanculturalpark.org/Exhibition.aspx?ID={}".format(e["ID"])
    } for e in exhibitionInfo]
    
    return exhibitionData
```

```python
exhibitionList=[]

for page in range(1,4):
    API = "https://www.songshanculturalpark.org/ExhibitionList.aspx?kind=0&type=1&p={}&q=get".format(page)
    exhibitionList.append(API)
    
exhibitionList
```

```python
allExhibitionData = []

for exhibationAPI in exhibitionList:
    exhibitionSoup = getExhibitionSoup(exhibationAPI)
    exhibitionData = getExhibitionInfo(exhibitionSoup)
    allExhibitionData.append(exhibitionData)
    
allExhibitionData
```
