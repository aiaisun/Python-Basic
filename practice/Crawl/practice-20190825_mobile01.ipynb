```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.mobile01.com/forumtopic.php?c=29"

headers = {
    "Referer": "https://www.mobile01.com/forumtopic.php?c=29",
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36",
        
}
res = requests.get(url,headers = headers)
soup = bs(res.text,"lxml")
print(soup)
```

```python
title = soup.select("div.l-listTable__tr > div.l-listTable__td > div.c-listTableTd > div.c-listTableTd__title > a")
titles = [a.text for a in title]
links = ["https://www.mobile01.com/"+a["href"] for a in title]
links
```
