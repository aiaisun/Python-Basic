```python
import requests, time, random
from bs4 import BeautifulSoup as bs

def getParsedData(soup,url):
    
#     分類
    catagory = soup.select("div.entry-crumbs")[0]
    catagoryAll = catagory.text.replace("  ",">")

    
#     標題 作者 時間
    news2 = soup.select("header.td-post-title")[0]
    title = news2.select("h1")[0].text
    author = news2.select("div.td-post-author-name")[0].text.replace(" ","").replace("-","").replace("/","").replace("\n","")
    time = news2.select("time")[0].text

#     內容    
    pTag = soup.select("span[itemprop='articleBody'] p")
    content=""
    for ele in pTag[:-1]:
        content+= ele.text
        content+="\n"
        
#     關鍵字        
    aTags = soup.select("div.td-post-source-tags li a")
    keyWords = [a.text for a in aTags]
    keyWordsList = "".join(keyWords)
    
#     更多連結
    moreTag = soup.select("div.relativeArticle_Reynold li a")
    moreList = [ {"Title": ele.text.strip(), "Link": ele["href"]} for ele in moreTag]

    data = {
        "url"      : url,
        "Title"    : title,
        "Catagory" : catagoryAll,
        "Author"   : author,
        "Time"     : time,
        "Content"  : content,
        "Key Words": keyWordsList,
        "More"     : moreList
        
    }
    return data



def saveFile(dataList,topic):
    with open("20190806-{}.csv".format(topic),"w",encoding="Big5") as outFiles:
        outFiles.write("標題,分類,作者,時間,連結,關鍵字\n")
        
        for data in dataList:
            try:
                content = ",".join([
                data["Title"],
                data["Catagory"],
                data["Author"],
                data["Time"],
                data["url"],
                data["Key Words"]
            ])
                outFiles.write(content)
                outFiles.write("\n")
            except:
                print(data["url"],"Error.")
    print(topic, "done.")


```

```python
import requests
from bs4 import BeautifulSoup as bs

# 分頁爬蟲
base_url = "https://www.nownews.com/cat/{}"
topic_url = "https://www.nownews.com/cat/{}/page/{}"
headers={
    "user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36"
}
target = ["entertainment","politics","society"]


for topic in target:
# #     準備url
    url = base_url.format(topic)
    topicLinks = []   
    for pageNum in range(1,4):
        urlPage = url+"/page/{}".format(pageNum)
#         print(urlPage)
            
# #     發request 取得頁面資料
        res = requests.get(urlPage, headers = headers)
        soup = bs(res.text,"lxml")   
            
# # #     取得文章列表
        article_links = [ele["href"] for ele in soup.select("div.td-module-container h3 a")]
#         print(soup.select("iv.td-module-container h3 a"))
        topicLinks += article_links
        print(urlPage, "is done")
    
    topicDataList = []
    for link in topicLinks:
        time.sleep(random.random()*2)
        url2 = link
        res2 = requests. get(url2, headers = headers)
        soup2 = bs(res2.text,"lxml")
        result = getParsedData(soup2,link)
        topicDataList.append(result)
        print("{} is done.".format(link))
    saveFile(topicDataList,topic)
print(len(topicLinks))      

    
# 存資料  

```
